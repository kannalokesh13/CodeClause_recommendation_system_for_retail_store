### Using The TfidfVectorizer To Generate The vectors for 'about_product' Column

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
tfv_matrix=tfv.fit_transform(data['about_product'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

### Using The Sigmoid Kernel To Identify the similarity between Products

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

### The Below Function Will Take A Product Id and Returns The Top 10 Most Similar Product ID's

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    product_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return data['product_id'].iloc[product_indices]
```
